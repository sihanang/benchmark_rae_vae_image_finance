```python
import pandas as pd
import numpy as np
import os
```

```python
# Getting the list of pointwise csv results
pointwise_names = [i for i in os.listdir('.') if 'mae.csv' in i]
```

```python
# Load them into dataframe before converting them to basis points
# after that round off to nearest 1 dp
!pip install xlsxwriter
with pd.ExcelWriter('pointwise_mae_consolidated.xlsx') as writer:
    for pointwise_name in pointwise_names:
        df = pd.read_csv(pointwise_name, index_col = 0)
        df = df * 10000
        df = df.round(1)
        df.index.name='Latent Dimensions'
        df.to_excel(writer, sheet_name = pointwise_name)

```
